```python
import pandas as pd
import numpy as np
```

## Task 1: Load and Initial Analysis
Read `ecommerce_sales.csv` into a DataFrame named `df`.
Store the shape of the DataFrame (a tuple) in `df_shape`.
Generate summary statistics for numerical columns using the `.describe()` method and store the result in `df_describe`.
Find the number of unique values in the `category` column and store it in `unique_categories_count`.

```python
df = pd.read_csv('ecommerce_sales.csv')
df_shape = df.shape
df_describe = df.describe()
unique_categories_count = df['category'].nunique()

print(f"Shape: {df_shape}")
print(f"Unique categories: {unique_categories_count}")
display(df_describe)
```

## Task 2: Cleaning Data
Identify rows that have missing values. Create a boolean mask indicating rows with any missing values and store it in `has_missing`.
Fill any missing values in the `discount_applied` column with `0.0`. Modify `df` in place.

```python
has_missing = df.isna().any(axis=1)
# Fill missing values in discount_applied column with 0.0
df['discount_applied'] = df['discount_applied'].fillna(0.0)
```

## Task 3: NumPy Array Operations
Extract the `price` and `quantity` columns as NumPy arrays.
Compute the total cost before discount (`price` multiplied by `quantity`) for each row using NumPy array multiplication, and store the resulting array in `gross_cost_array`.

```python
price_array = df['price'].values
quantity_array = df['quantity'].values
gross_cost_array = price_array * quantity_array
print(gross_cost_array[:5])
```

## Task 4: Feature Engineering
Create a new column in `df` called `net_revenue`.
The formula is: `(price * quantity) * (1 - discount_applied) + shipping_cost`

```python
df['net_revenue'] = (df['price'] * df['quantity']) * (1 - df['discount_applied']) + df['shipping_cost']
df[['price', 'quantity', 'discount_applied', 'shipping_cost', 'net_revenue']].head()

```

## Task 5: GroupBy Analysis
Group the data by `category`.
Calculate the **sum** of `net_revenue` and the **mean** of `discount_applied` for each category.
Store the resulting DataFrame in `category_stats`.

```python
category_stats = df.groupby('category').agg(
    sum_net_revenue=('net_revenue', 'sum'),
    mean_discount_applied=('discount_applied', 'mean')
)
category_stats
```

## Task 6: Pivot Table
Create a pivot table showing the total (sum) `net_revenue` for each `category` (index/rows) broken down by `payment_method` (columns).
Store the pivot table DataFrame in `payment_pivot`. Fill any missing values in the pivot table with `0`.

```python
payment_pivot = pd.pivot_table(
    df,
    values='net_revenue',
    index='category',
    columns='payment_method',
    aggfunc='sum',
    fill_value=0
)
payment_pivot
```

## Task 7: Boolean Indexing & Sorting
Find all orders where `shipping_cost` is greater than 10.0 **AND** `net_revenue` is greater than 50.0.
Sort these orders by `net_revenue` in **descending** order.
Store the top 5 resulting rows in a new DataFrame called `premium_orders`.

```python
premium_orders = df[(df['shipping_cost'] > 10.0) & (df['net_revenue'] > 50.0)]
premium_orders = premium_orders.sort_values('net_revenue', ascending=False).head(5)
premium_orders
```
